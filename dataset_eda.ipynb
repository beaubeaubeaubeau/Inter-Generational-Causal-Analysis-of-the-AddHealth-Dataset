# Wave 1

```python
import pandas as pd
import os

w1_dir = 'data/W1'

print("Loading w1context...")
w1_context = pd.read_sas(os.path.join(w1_dir, 'w1context.sas7bdat'))

print("Loading w1inhome...")
w1_inhome = pd.read_sas(os.path.join(w1_dir, 'w1inhome.sas7bdat'))

print("Loading w1network...")
w1_network = pd.read_sas(os.path.join(w1_dir, 'w1network.sas7bdat'))

print("Loading w1weight...")
w1_weight = pd.read_sas(os.path.join(w1_dir, 'w1weight.sas7bdat'))

print("Context shape:", w1_context.shape)
print("In-home shape:", w1_inhome.shape)
print("Network shape:", w1_network.shape)
print("Weight shape:", w1_weight.shape)
```

### w1_context data
Neighborhood, community, and school-level environmental data collected during Wave 1 (1994-1995), primarily drawn from the 1990 U.S. Census mapped to the adolescent's geographic location. Variables include block-level poverty rates, median income, unemployment, and racial composition (`BST90P...` variables).

Note: Large values (e.g. 999998.0, 9998.0) is used to encode missing values.

```python
w1_context.head()
```

### w1_inhome data
Contains the survey data administered to adolescents and their parents in their homes. Adolescent data have the prefix M1 whilst parrent data have the prefix PA or PB.

Note: Reference the Wave 1 Codebook to translate cryptic variable names and their values. If a student answered "No" to drinking alcohol, all subsequent questions about alcohol are skipped (coded w/ numbers unding in 7, 97, 997, etc)

```python
w1_inhome.head()
```

### w1_network data
Pre-calculated structural graph-theory metrics based on the adolescent's friendship nominations at school. Variables include popularity (Indegree), sociability (Outdegree), and network centrality. 

Note: captures relationship intraschool, rather than interschool

```python
w1_network.head()
```

### w1_weights data
Statistical weights used to correct for AddHealth's complex, stratified sampling design. Because AddHealth intentionally oversampled specific minority groups and disabled students, the raw data is not naturally representative of the 1994 U.S. adolescent population. Use weights to generalize results to popualtion

```python
w1_weight.head()
```

# W5

```python
w5_dir = 'data/W5'

print("Loading W5 Anthropometrics...")
w5_anthro = pd.read_sas(os.path.join(w5_dir, 'panthro5.xpt'), format='xport')

print("Loading W5 Cardiovascular...")
w5_cardio = pd.read_sas(os.path.join(w5_dir, 'pcardio5.xpt'), format='xport')

print("Loading W5 Main Survey...")
w5_wave5 = pd.read_sas(os.path.join(w5_dir, 'pwave5.xpt'), format='xport')
```

### w5_anthro data
Biological and physical measurements taken when the respondents were adults (aged 32-42) during the 2016-2018 Wave 5 visits. Includes variables like adult BMI, waist circumference, height, and weight. Highly useful for defining long-term metabolic or physical health outcomes.

Note: Women who were pregnant during the Wave 5 visit often have their anthropometric variables safely coded as missing or skipped to avoid throwing off BMI and weight distributions. 

```python
w5_anthro.head()
```

### w5_cardio data
Adult cardiovascular health markers, primarily focusing on Systolic and Diastolic blood pressure, as well as resting heart rate. These act as direct, objective biomarkers for cardiovascular risk, hypertension, and stress.

```python
w5_cardio.head()
```

### w5_wave5 data
The adult counterpart to the `w1_inhome` survey. It contains self-reported data on the respondent's adult life: socioeconomic status (income, wealth, education), marital status, mental health, and substance use. This is where you look for inter-generational mobility (e.g., did a child from a low-income Wave 1 household achieve a high-income adult status?).

Note: A significant portion of Wave 1 adolescents dropped out of the study by Wave 5. If the people who dropped out are systematically different from those who stayed (e.g., lower income), the causal analysis could be biased. Because of this attrition, we can't just use the w1_weight for a W1 → W5 analysis. We must use specific Wave 5 longitudinal weights (usually provided in a W5 weights file) that adjust for the non-random dropout rate over the 20-year gap.

```python
w5_wave5.head()
```

# Merging W1 and W5

```python
dataframes = [w1_context, w1_inhome, w1_network, w1_weight, w5_anthro, w5_cardio, w5_wave5]
for df in dataframes:
    if 'AID' in df.columns and pd.api.types.is_object_dtype(df['AID']):
        df['AID'] = df['AID'].apply(lambda x: x.decode('utf-8') if isinstance(x, bytes) else x)

w1_w5_df = pd.merge(left=w1_inhome, right=w5_anthro, on='AID', how='inner')

print(f"Original Wave 1 participants: {len(w1_inhome)}")
print(f"Participants remaining in Wave 5 (Anthro): {len(w1_w5_df)}")
print(f"Attrition rate: {1 - len(w1_w5_df)/len(w1_inhome):.1%}")
```

```python
# This section TBC later on; identify more variables

causal_df = w1_w5_df[[
    'AID', 
    'PA55',     # Treatment (W1): Parent Household Income
    'H1GI1M',   # Confounder (W1): Month of Birth
    'BIO_SEX',  # Confounder (W1): Biological Sex
    'H5BMI',    # Outcome (W5 Anthropometrics): Adult BMI
]].copy()

causal_df.rename(columns={
    'PA55': 'w1_parent_income',
    'H1GI1M': 'w1_birth_month',
    'BIO_SEX': 'baseline_sex',
    'H5BMI': 'w5_adult_bmi'
}, inplace=True)

causal_df.head()
```

```python
import matplotlib.pyplot as plt
import numpy as np

causal_df['w1_parent_income'] = causal_df['w1_parent_income'].replace(9996.0, np.nan)

plt.hist(causal_df['w1_parent_income'], bins=50)
plt.title("Wave 1 Parent Income Distribution")
plt.show()
```

```python
print(causal_df['baseline_sex'].unique())

print(causal_df['baseline_sex'].value_counts(dropna=False))
```

```python
causal_df['w5_adult_bmi'] = causal_df['w5_adult_bmi'].replace(9999.0, np.nan)
causal_df['w5_adult_bmi'] = causal_df['w5_adult_bmi'].replace(9997.0, np.nan)
causal_df['w5_adult_bmi'] = causal_df['w5_adult_bmi'].replace(9994.0, np.nan)

plt.hist(causal_df['w5_adult_bmi'], bins=50)
plt.title("Wave 5 BMI Distribution")
plt.show()
```
